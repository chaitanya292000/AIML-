```python
# in dropna , for removing all column by using how='all' ,  we will make toy axis as NaN
df = pd.DataFrame({'name':['Alfred','Batman','Catwoman'],
                   'toy':[np.nan,np.nan,np.NaN],
                   'born':[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]
                  
                  })
```

```python
df.head()
```

```python
# drops when their is NaN in dataframe
df.dropna()
```

```python
df.dropna(how="all",axis=1)
```

```python
df.head()
```

```python
# for dropping dupliates , we will make alfred two times
df = pd.DataFrame({'name':['Alfred','Batman','Alfred'],
                   'toy':[np.nan,'Batmobile','Bullwhip'],
                   'born':[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]
                  
                  })
```

```python
# df.drop_duplicates(subset=['name'],keep='first') # default
# df.drop_duplicates(subset=['name'],keep='last') # for last element
df.drop_duplicates(subset=['name'],keep=False)   # removes both the records 
```
